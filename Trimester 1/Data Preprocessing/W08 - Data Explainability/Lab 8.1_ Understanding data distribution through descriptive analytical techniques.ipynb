## Lab Summary: Understanding Data Distribution through Descriptive Analysis

This lab focuses on applying descriptive analytical techniques to explore and understand the distribution of data within a housing dataset. Specifically, the analysis centers on the 'LotArea' and 'SalePrice' columns to gain insights into their central tendency, dispersion, shape, and potential outliers.

Key descriptive statistics calculated include:

- **Measures of Central Tendency:** Mean, Median, and Mode to identify the typical values in the datasets.
- **Measures of Dispersion:** Range, Quartiles, Interquartile Range (IQR), Standard Deviation, and Variance to understand the spread and variability of the data.
- **Measures of Shape:** Skewness and Kurtosis to assess the asymmetry and tailedness of the distributions, highlighting the presence of outliers.
- **Relationship between Variables:** Correlation Coefficient between 'LotArea' and 'SalePrice' to examine the linear relationship between these two variables.
- **Data Quality Assessment:** Analysis of missing values in the 'Fence' column to identify data completeness issues.

By utilizing these descriptive statistics, the lab provides a comprehensive overview of the data's characteristics, which is crucial for subsequent data cleaning, transformation, and modeling steps in a data analysis workflow.

```python
# Import required libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataset file into memory
DATASET_PATH = "./housing.csv"
df = pd.read_csv(DATASET_PATH)
df
```

```python
# Displaying data types of all columns in dataframe
df.dtypes
```

```python
# Performing various kinds of descriptive analysis
```

```python
# Calculate and print the mean (average) of the 'LotArea' column.
print("Mean lot area:", df["LotArea"].mean())
# The mean lot area is 10516.83. This value is higher than the median, which
# shows it is being pulled up by a few large properties.

# Calculate and print the median (middle value) of the 'LotArea' column.
#
# This is useful when the data is skewed, as it's not affected by extremely
# large or small outliers.
print("Median lot area:", df["LotArea"].median())
# The median lot area is 9478.5. This is a more representative value for the
# typical lot size, as it is not affected by outliers.

# Calculate and print the mode (most frequent value) of the 'LotArea' column.
print("Mode lot area:", df["LotArea"].mode())
# The mode is 7200. This is the most frequently occurring lot area in the dataset.

# Adding an empty line for better readability in the output.
print()

# Calculate and print the smallest lot area.
print("Smallest lot area:", df["LotArea"].min())
# The smallest lot area is 1300.

# Calculate and print the largest lot area.
print("Largest lot area:", df["LotArea"].max())
# The largest lot area is 215245. This value is an extreme outlier that
# significantly affects the mean.

# Calculate and print the total of all lot areas combined.
print("Total lot area:", df["LotArea"].sum())
# The total combined lot area is 15354569.

# Calculate and print the range (difference between the largest and smallest
# values) of the 'LotArea' column.
print("Range:", df["LotArea"].max() - df["LotArea"].min())
# The range of 213945 shows the very large spread between the smallest and
# largest lot areas.
```

```python
# Calculate and print the first quartile (25th percentile) of the 'LotArea'
# column.
#
# This value represents the point below which 25% of the data falls.
print("\nFirst-Quartile of LotArea:")
print(df["LotArea"].quantile(0.25))
# A value of 7553.5 shows that 25% of the lots are 7553.5 square feet or smaller.

# Calculate and print the third quartile (75th percentile) of the 'LotArea' column.
#
# This value represents the point below which 75% of the data falls (or above
# which 25% of the data falls).
print("\nThird-Quartile of LotArea:")
print(df["LotArea"].quantile(0.75))
# A value of 11601.5 shows that 75% of the lots are 11601.5 square feet or smaller.

# Calculate and print the Interquartile Range (IQR), which is the difference
# between the third and first quartiles.
#
# The IQR measures the spread of the middle 50% of the data and is resistant to outliers.
print("\nInter-Quartile of LotArea:")
print(df["LotArea"].quantile(0.75) - df["LotArea"].quantile(0.25))
# An IQR of 4048.0 shows that the middle 50% of the lot areas are spread across
# a range of 4048 square feet.
```

```python
# Defining a function to calculate the Coefficient of Variation (CV).
#
# The formula is (Standard Deviation / Mean) * 100, which gives a percentage.
#
# We use 'ddof=1' to calculate the sample standard deviation.
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100

# Calculate and print the standard deviation of the 'LotArea' column.
print("Standard Deviation of lot area column:", df["LotArea"].std())
# A standard deviation of 9981.26 shows that lot areas are very spread out from
# the mean.

# Calculate and print the variance of the 'LotArea' column.
print("Variance of lot area column:", df["LotArea"].var())
# The large variance of 99625649.65 also confirms the wide spread of the lot
# area data.

# Calculate and print the correlation coefficient between 'LotArea' and
# 'SalePrice'.
print("Co-relation of lot area with sales price:", df["LotArea"].corr(df["SalePrice"]))
# A correlation of 0.26 indicates a weak positive linear relationship. Larger
# lots tend to have higher prices, but other factors are more influential.

# Add an empty line for better readability.
print()

# Calculate and print the standard error of the mean (SEM) for the 'Lot Area'
# column.
print("Standard error of mean of 'Lot Area' column:", df["LotArea"].sem())
# This value of 261.22 estimates how close the sample average is to the true
# population average.

# Calculate and print the Coefficient of Variation (CV) for both 'LotArea' and
# 'SalePrice'.
print("Co-efficient of variance for LotArea and SalePrice column:\n")
print(df[["LotArea", "SalePrice"]].apply(cv))
# The CV for LotArea is much higher than for SalePrice, meaning lot sizes are
# relatively more dispersed than sales prices.

# Add an empty line for better readability.
print()

# Count and print the number of missing (NaN) values in the 'Fence' column.
print("Number of missing values in 'Fence' column:", df["Fence"].isna().sum())
# This shows that 1179 values are missing, indicating a large portion of the
# 'Fence' data is incomplete.

# Count and print the number of non-missing values in the 'Fence' column.
print("Number of distinct values in 'Fence' column:", df["Fence"].count())
# This count confirms there are only 281 valid entries for the 'Fence' column.
```

```python
# Calculate the sum of the squared values of the 'LotArea' column.
#
# It first squares each individual value, then adds them all up.
print("Sum of squares of Lot Area:", (df["LotArea"] ** 2).sum())
# The result is a large number as expected, serving as a building block for
# variance calculation.

# Calculate the skewness of the 'LotArea' column.
#
# Skewness measures the asymmetry of the data distribution. A positive value
# indicates that the data has a long tail to the right, with a few high values
# stretching the distribution in that direction.
print("Skewness of Lot Area column:", df["LotArea"].skew())
# A value of 12.21 indicates the data is highly positively skewed due to a few
# extreme outliers.

# Calculate the kurtosis of the 'LotArea' column.
#
# Kurtosis measures the "tailedness" of the distribution, which tells you about
# the presence of extreme outliers. A positive value suggests that the
# distribution has heavier tails than a normal distribution, meaning outliers
# are more frequent.
print("Kurtosis of Lot Area column:", df["LotArea"].kurtosis())
# A value of 203.24 indicates a very high kurtosis, confirming
# the presence of severe outliers and heavy tails.
```
