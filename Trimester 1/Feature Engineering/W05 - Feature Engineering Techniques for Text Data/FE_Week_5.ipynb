# Sample Dataset


```python
documents = [
    "This product is amazing",
    "Very bad quality",
    "Not worth the money"
]

labels = ["Positive", "Negative", "Negative"]
```

# Basic Preprocessing (Lowercase)

```python
documents = [doc.lower() for doc in documents]
documents
```

## Remove Stopwords

# Bag of Words (Unigrams)

```python
from sklearn.feature_extraction.text import CountVectorizer

## Remove Stopwords
custom_stopwords = ["this", "is", "very", "the"]

vectorizer = CountVectorizer(stop_words=custom_stopwords)
X_bow = vectorizer.fit_transform(documents)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", X_bow.toarray())
```

# Bag of Words with Bigrams

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "This product is amazing",
    "Very bad quality",
    "Not worth the money"
]

custom_stopwords = ["this", "is", "very", "the"]

vectorizer_bigram = CountVectorizer(
    stop_words=custom_stopwords,
    ngram_range=(1, 2)
)

X_bigram = vectorizer_bigram.fit_transform(documents)

print("Vocabulary:", vectorizer_bigram.get_feature_names_out())
print("BoW (Unigram + Bigram):\n", X_bigram.toarray())

```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "This product is amazing",
    "Very bad quality",
    "Not worth the money"
]

custom_stopwords = ["this", "is", "very", "the"]

tfidf = TfidfVectorizer(
    stop_words=custom_stopwords,
    ngram_range=(1, 2)   # Unigrams + Bigrams
)

X_tfidf = tfidf.fit_transform(documents)

print("Vocabulary:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", X_tfidf.toarray())

```

# Simple Classification Model
## Predicting the Class label for the same training data documents/sentences

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_tfidf, labels)

predictions = model.predict(X_tfidf)
print("Predictions:", predictions)
```

# Showing Feature Importance

```python
import pandas as pd

feature_names = tfidf.get_feature_names_out()
coefficients = model.coef_[0]  ## The importance values are coming from LogisticRegression model in scikit-learn
                               ## model.coef_  returns the learned coefficients (weights) of the linear model.
importance = pd.DataFrame({
    "Feature": feature_names,
    "Weight": coefficients
}).sort_values(by="Weight", ascending=False)

print(importance)
```
