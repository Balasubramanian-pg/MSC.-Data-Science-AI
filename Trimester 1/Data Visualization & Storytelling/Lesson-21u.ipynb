## Creating a Plot: An Interactive Example
The cell below contains the code to generate a simple sine wave. You will also see several lines of code that are "commented out" with a `#` symbol.

**Your task is to uncomment these lines one by one (by removing the `#`) and re-running the cell to see how each line changes the plot.**

```python
# --- 1. Prepare Data ---
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x) # Data for the second plot

# --- 2. Create Plot ---
fig, ax = plt.subplots()

# --- 3. Plot Data ---
# Initial plot
ax.plot(x, y, label='Sine')

# ==> UNCOMMENT the line below to plot a cosine wave on the same graph!
# ax.plot(x, y2, color='red', linestyle='--', label='Cosine')


# --- 4. Customize Plot ---
ax.set_title("Simple Sine Wave")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")

# ==> UNCOMMENT the line below to add a grid for better readability.
# ax.grid(True)

# ==> UNCOMMENT the line below to add a legend.
# ax.legend()


# --- 5. Show Plot ---
plt.show()
```

## Final Exercise: Your Turn!

Now it's time to combine what you've learned. Your task is to plot a simple linear function.

**Task: Plot the function y = 2x + 5**

1.  Generate x-values from -5 to 5.
2.  Calculate the corresponding y-values for the function `y = 2x + 5`.
3.  Plot the function as a `green line`.
4.  Add a title ("Linear Function: y = 2x + 5") and labels for the x and y axes.
5.  Display the plot.

```python
# Solution
x = np.linspace(-5, 5, 100)
y = 2 * x + 5

fig, ax = plt.subplots()
ax.plot(x, y, color='green')

ax.set_title("Linear Function: y = 2x + 5")
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")
ax.grid(True)

plt.show()
```
