# Decorating Visuals in Bokeh

## Introduction

This notebook explores how Bokeh handles:

- plot styling
- glyph customization
- legends
- axes
- themes
- color systems
- color mapping
- title customization

The core architectural idea is:

```text
Everything in Bokeh is an editable object.
```

This includes:

- figures
- glyphs
- axes
- titles
- legends
- grids
- color bars

Bokeh behaves more like a UI framework than a traditional static plotting library.

# 4. Bokeh vs Matplotlib

| Feature | Bokeh | Matplotlib |
|---|---|---|
| Rendering | Interactive HTML | Static image |
| Display | `show()` | `plt.show()` |
| Architecture | Object-oriented UI system | Primarily plotting system |
| Glyphs | Explicit glyph model | Implicit drawing model |
| Interactivity | Native | Limited by default |

# 7. Basic Categorical Bar Chart

````python
fruits = [
    "Apple",
    "Mango",
    "Orange",
    "Banana"
]

counts = [10, 25, 18, 30]

p = figure(
    x_range=fruits,
    height=350,
    width=700,
    title="Fruit Counts"
)

p.vbar(
    x=fruits,
    top=counts,
    width=0.5,
    color="indigo"
)

show(p)
````

# 8. Why `x_range` Matters

`x_range=fruits` converts the x-axis from:

```text
continuous
```

to:

```text
categorical
```

Visual intuition:

```text
Continuous Axis:
1 ---- 2 ---- 3 ---- 4

Categorical Axis:
Apple | Mango | Orange | Banana
```

# 9. RGB and RGBA Examples

````python
rgb_plot = figure(
    x_range=fruits,
    height=350,
    title="RGB Example"
)

rgb_plot.vbar(
    x=fruits,
    top=counts,
    width=0.5,
    color=(150, 10, 10)
)

show(rgb_plot)
````

````python
rgba_plot = figure(
    x_range=fruits,
    height=350,
    title="RGBA Example"
)

rgba_plot.vbar(
    x=fruits,
    top=counts,
    width=0.5,
    color=(100, 100, 100, 0.4)
)

show(rgba_plot)
````

# 13. Visual Properties Overview

# 14. Figure Creation

````python
visual_properties_plot = figure(
    height=350,
    width=700,
    title="Domestic Freight Mail"
)
````

# 15. Title Styling

````python
visual_properties_plot.title.text_font_size = "18pt"
visual_properties_plot.title.text_color = "navy"
visual_properties_plot.title.text_font_style = "bold"
````

# 16. Line Styling

| Property | Meaning |
|---|---|
| line_width | line thickness |
| line_color | line color |
| line_alpha | transparency |
| line_dash | dash pattern |

Dash styles:

```text
Solid:
──────────

Dashed:
- - - - - -

Dotted:
..........
```

````python
x = monthly_values_df.index
freight = monthly_values_df["freight"]

visual_properties_plot.line(
    x,
    freight,
    line_width=3,
    line_color="blue",
    line_alpha=0.7,
    line_dash="dashed"
)

show(visual_properties_plot)
````

# 17. Fill and Hatch Properties

````python
fill_plot = figure(
    height=350,
    width=700,
    title="Fill and Hatch Example"
)

fill_plot.vbar(
    x=list(range(12)),
    top=monthly_values_df["freight"],

    width=0.8,

    fill_color="orange",
    fill_alpha=0.6,

    line_color="black",
    line_width=2,

    hatch_pattern="/",
    hatch_color="red"
)

show(fill_plot)
````

# 18. Fill vs Line Properties

| Property | Controls |
|---|---|
| fill_color | interior |
| line_color | border |
| hatch_pattern | internal pattern |
| hatch_color | pattern color |

# 19. Plot-Level Styling

````python
plot = figure()

plot.height = 300
plot.width = 700

plot.outline_line_color = "navy"
plot.outline_line_width = 2
plot.outline_line_alpha = 0.5

plot.background_fill_color = "lightblue"

plot.xgrid.grid_line_color = None
plot.ygrid.grid_line_color = None

plot.scatter(
    [1,2,3,4,5],
    [2,5,7,8,2],
    size=10
)

show(plot)
````

# 20. Glyph Objects

# 21. Circle Glyph Example

````python
glyph_plot = figure(height=300)

circle = glyph_plot.circle(
    x=[1,2,3],
    y=[2,5,8],

    size=15,

    fill_color="yellow",
    line_color="red"
)

show(glyph_plot)
````

# 22. Dynamic Glyph Modification

````python
editable_plot = figure(height=300)

circle = editable_plot.circle(
    [1,2,3],
    [2,5,8],

    size=20,
    fill_color="yellow",
    line_color="red"
)

# Modify later
circle.glyph.fill_color = "green"
circle.glyph.line_color = "blue"
circle.glyph.line_width = 3

show(editable_plot)
````

# 23. Why Editable Glyphs Matter

Bokeh maintains live visual objects instead of static rendering.

# 24. Axis Styling

````python
axis_plot = figure(
    height=300,
    width=600
)

x = [1,2,3,4,5]
y = [2,5,3,8,6]

axis_plot.line(
    x,
    y,
    line_color="purple",
    line_width=3
)

axis_plot.xaxis.axis_label = "Month"
axis_plot.yaxis.axis_label = "Freight"

axis_plot.xaxis.axis_line_color = "red"
axis_plot.yaxis.axis_line_color = "blue"

axis_plot.xaxis.major_tick_line_color = "red"
axis_plot.xaxis.minor_tick_line_color = "orange"

axis_plot.yaxis.major_label_text_font_size = "1.2em"

axis_plot.xgrid.grid_line_color = "gray"

show(axis_plot)
````

# 26. Label Orientation

Long categorical labels often overlap.

Bokeh solves this using rotation in radians.

| Value | Meaning |
|---|---|
| `0` | horizontal |
| `0.78` | diagonal |
| `1.57` | vertical |

````python
carrier_names = [
    "Delta Airlines",
    "American Airlines",
    "United Airlines",
    "Lufthansa",
    "Emirates",
    "Qatar Airways",
    "Singapore Airlines",
    "Air India"
]

passengers = [
    14000000,
    12000000,
    11000000,
    10000000,
    9500000,
    9200000,
    9000000,
    8500000
]

rotation_plot = figure(
    x_range=carrier_names,
    height=400,
    width=800,
    title="Top Carriers by Passengers"
)

rotation_plot.vbar(
    x=carrier_names,
    top=passengers,
    width=0.5
)

rotation_plot.xaxis.major_label_orientation = 0.8

show(rotation_plot)
````

# 27. Tick Formatting

Raw large numbers are cognitively expensive.

Formatting improves readability.

| Format | Example |
|---|---|
| `0` | 14000000 |
| `0,0` | 14,000,000 |
| `0.0a` | 14.0m |

````python
rotation_plot.yaxis.formatter = NumeralTickFormatter(
    format="0.0a"
)

show(rotation_plot)
````

# 30. Legends

````python
legend_plot = figure(
    title="Legend Example",
    height=400
)

x = [1,2,3,4,5]

y1 = [2,4,6,8,10]
y2 = [1,3,2,5,4]

legend_plot.line(
    x,
    y1,

    legend_label="Temperature",

    line_color="blue",
    line_width=2
)

legend_plot.scatter(
    x,
    y2,

    legend_label="Objects",

    size=10,
    color="red"
)

legend_plot.legend.location = "top_left"
legend_plot.legend.title = "Observations"

show(legend_plot)
````

# 31. Advanced Legend Customization

````python
legend_plot.legend.label_text_color = "navy"

legend_plot.legend.label_text_font = "Times New Roman"

legend_plot.legend.label_text_font_style = "italic"

legend_plot.legend.border_line_width = 3
legend_plot.legend.border_line_color = "navy"
legend_plot.legend.border_line_alpha = 0.8

legend_plot.legend.background_fill_color = "navy"
legend_plot.legend.background_fill_alpha = 0.2

show(legend_plot)
````

# 35. Color Mapping

Color mapping converts:

```text
numeric values
      →
colors
```

Used heavily in:

- heatmaps
- ML feature visualization
- scientific visualization
- geospatial systems

The example dataset uses the parabola:

$$
y = x^2
$$

````python
x = list(range(-32, 33))
y = [i**2 for i in x]

source = ColumnDataSource(data=dict(x=x, y=y))
````

# 36. Linear Color Mapping

````python
mapper = linear_cmap(
    field_name='y',
    palette=Turbo256,
    low=min(y),
    high=max(y)
)
````

# 37. Scatter Plot with Color Encoding

````python
mapper_plot = figure(
    title="Linear Color Mapping Example",
    height=400,
    width=700
)

mapper_plot.scatter(
    'x',
    'y',

    source=source,

    color=mapper,
    size=10
)

show(mapper_plot)
````

# 38. Color Bars

Color bars are continuous legends.

They explain:

```text
which color corresponds
to which value
```

````python
color_bar = ColorBar(
    color_mapper=mapper['transform']
)

mapper_plot.add_layout(
    color_bar,
    'right'
)

show(mapper_plot)
````

# 39. Themes

Themes act similarly to CSS in web development.

| Theme | Style |
|---|---|

# 41. Common Beginner Mistakes

## Overstyling

Too many:

- colors
- thick borders
- hatch patterns
- fonts

creates clutter.

## Forgetting `x_range`

Categorical axes require:

```python
x_range=categories
```

## Raw Large Numbers

Avoid:

```text
15438293
```

Prefer:

```text
15.4M
```

## Decorative Color Usage

Color should encode meaning, not decoration.

## Excessive Label Rotation

Diagonal labels are often more readable than fully vertical labels.
