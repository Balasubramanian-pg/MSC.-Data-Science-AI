# Bokeh Glyphs, DataFrames, Scatter Plots, and Categorical Visualization

## Introduction

Bokeh is an interactive visualization framework built around one central architectural idea:

```text
Data
  → Glyphs
      → Renderers
          → Figure
              → Browser
```

Every visual object in Bokeh is treated as a glyph.

Examples:

- line
- scatter
- circle
- square
- rectangle
- bar
- triangle

Bokeh internally converts these glyphs into browser-rendered interactive objects.

# 1. Import Required Libraries

````python
import pandas as pd
import numpy as np

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
````

# 2. Initialize Notebook Rendering

````python
output_notebook()
````

## Why `output_notebook()` Matters

Bokeh renders visualizations inside the browser using HTML and JavaScript.

This function initializes the browser rendering engine within Jupyter or Google Colab.

```mermaid
flowchart LR

A[Notebook]
    --> B[output_notebook()]
    --> C[Load BokehJS]
    --> D[Interactive Visualization]
```

# 5. Creating a Basic Line Plot

````python
# Generate data
x = list(range(1, 51))
y = [i**2 for i in x]
````

## Mathematical Relationship

The plotted function is:

```text
y = x²
```

````python
parabola_plot = figure(
    height=400,
    width=700,
    title='Parabola Plot',
    x_axis_label='X Values',
    y_axis_label='Y Values'
)

parabola_plot.line(
    x,
    y,
    line_width=3,
    color='blue',
    legend_label='y = x²'
)

show(parabola_plot)
````

# 6. Why `show()` is Mandatory

Unlike Matplotlib, Bokeh does not automatically render plots.

`show()` converts the Python figure object into browser-rendered HTML and JavaScript.

## Bokeh vs Matplotlib

| Matplotlib | Bokeh |
|---|---|
| Static rendering | Interactive browser rendering |
| `plt.plot()` | `plot.line()` |
| Automatic notebook display | Explicit `show()` required |
| Image-centric | Application-centric |

# 9. Bokeh with Pandas DataFrames

Real-world analytics systems primarily use Pandas DataFrames.

Bokeh integrates naturally with Pandas-based workflows.

## Create Airline Passenger Dataset

````python
months = list(range(1, 13))

passengers = [
    35.1, 36.2, 38.4, 40.1,
    42.3, 45.0, 47.8, 49.5,
    50.1, 51.2, 53.0, 54.6
]
````

## Create DataFrame

````python
monthly_df = pd.DataFrame(
    data={'Passengers': passengers},
    index=months
)

monthly_df.index.name = 'Month'

monthly_df.head()
````

## Data Pipeline Architecture

```mermaid
flowchart LR

A[Raw Data]
    --> B[Pandas DataFrame]
    --> C[Bokeh Figure]
    --> D[Interactive Visualization]
```

## Plot DataFrame Data

````python
airline_plot = figure(
    height=400,
    width=700,
    title='Domestic Airline Passengers in 2021',
    x_axis_label='Month',
    y_axis_label='Passengers (Millions)'
)

airline_plot.line(
    monthly_df.index,
    monthly_df['Passengers'],
    line_width=3,
    line_color='blue',
    legend_label='Passengers'
)

show(airline_plot)
````

## Customized Scatter Plot

````python
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [6, 7, 8, 4, 5]

custom_scatter = figure(
    width=500,
    height=300,
    title='Customized Scatter Plot'
)

custom_scatter.scatter(
    x_scatter,
    y_scatter,
    marker='triangle',
    size=20,
    color='green',
    alpha=0.6
)

show(custom_scatter)
````

# 13. Combining Multiple Glyphs

Multiple glyphs can coexist on the same figure.

This enables:

- overlays
- trend + detail
- comparisons
- anomaly highlighting
- layered storytelling


## Combined Line + Scatter Plot

````python
x_combined = [1, 2, 3, 4, 5]
y_combined = [6, 7, 8, 4, 5]

combined_plot = figure(
    width=500,
    height=300,
    title='Line and Scatter Combined'
)

# Line glyph
combined_plot.line(
    x_combined,
    y_combined,
    line_width=2,
    color='blue'
)

# Scatter glyph
combined_plot.scatter(
    x_combined,
    y_combined,
    size=12,
    color='red'
)

show(combined_plot)
````

## Multi-Glyph Architecture

```mermaid
flowchart TD

A[Figure]
    --> B[Line Glyph]
    --> C[Scatter Glyph]
    --> D[Square Glyph]

B --> E[Combined Interactive Visualization]
C --> E
D --> E
```

# 14. Layering Multiple Scatter Glyphs

````python
layered_plot = figure(
    width=500,
    height=300,
    title='Layered Scatter Glyphs'
)

layered_plot.scatter(
    x_combined,
    y_combined,
    marker='triangle',
    size=20,
    color='yellow',
    alpha=0.6
)

layered_plot.scatter(
    x_combined,
    y_combined,
    marker='square',
    size=10,
    color='pink',
    alpha=0.8
)

show(layered_plot)
````

# 15. Categorical Data Visualization

Categorical variables contain labels rather than continuous numerical coordinates.

Examples:

- products
- fruits
- regions
- weekdays
- smoker/non-smoker


## Continuous vs Categorical Axes

| Continuous Axis | Categorical Axis |
|---|---|
| Numerical distances matter | Symbolic ordering matters |
| Used in line/scatter plots | Used in bar charts |
| Example: 1,2,3,4 | Example: Apples, Pears |


# 16. Creating a Bar Chart

## Create Fruit Dataset

````python
fruits = [
    'Apples',
    'Pears',
    'Plums',
    'Grapes',
    'Nectarines',
    'Strawberries'
]

counts = [5, 3, 4, 2, 6, 7]
````

## Why `x_range` Matters

`x_range` tells Bokeh:

```text
Treat the x-axis as categorical.
```

Without `x_range`, Bokeh expects numerical coordinates.

## Numeric vs Categorical Coordinate Systems

### Numeric Axis

```text
0 --- 1 --- 2 --- 3
```

### Categorical Axis

```text
Apple   Pear   Plum
```

## Create Categorical Figure

````python
bar_plot = figure(
    x_range=fruits,
    height=350,
    width=700,
    title='Fruit Counts',
    x_axis_label='Fruit Type',
    y_axis_label='Count'
)
````

## Add Vertical Bar Glyph

````python
bar_plot.vbar(
    x=fruits,
    top=counts,
    width=0.6,
    color='firebrick'
)

show(bar_plot)
````

# 20. Common Beginner Mistakes

## Forgetting `show()`

Plots will not render.

## Forgetting `output_notebook()`

Notebook rendering may fail.

## Missing `x_range` for Categories

Categorical plots may behave incorrectly.

## Mismatched Data Lengths

X and Y arrays must have equal lengths.

## Excessive Glyph Layers

Too many glyphs increase browser overhead.

# 21. Advanced Architectural Insight

Every glyph in Bokeh becomes:

- a renderer object
- a JavaScript object
- a browser-managed interactive element

This enables:

- hover interactions
- streaming updates
- dashboards
- callbacks
- linked visualizations

But also introduces:

- browser memory overhead
- rendering limitations for massive datasets
