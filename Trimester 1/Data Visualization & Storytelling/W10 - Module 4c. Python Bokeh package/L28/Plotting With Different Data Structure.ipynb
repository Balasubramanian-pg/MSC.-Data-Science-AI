# 3. Continuous vs Categorical Axes

## Continuous Axis

Used for:

- line plots
- scatter plots
- time series

Example:

```text
1, 2, 3, 4, 5
```

Distances are interpreted numerically.

## Categorical Axis

Used for:

- bar charts
- grouped comparisons
- frequency analysis

Example:

```text
["Apple", "Pear", "Plum"]
```

Spacing is symbolic rather than numerical.

# 4. Visual Mental Model

## Continuous Axis

```text
0 --- 1 --- 2 --- 3
```

Distance matters.

## Categorical Axis

```text
Apple   Pear   Plum
```

Order matters.
Distance does not.

# 5. Why Bar Charts Exist

Bar charts are designed for:

- category comparison

- aggregated metrics

Good for:

Bad for:

# 6. Create Fruit Dataset

````python
fruits = [
    "Apples",
    "Pears",
    "Plums",
    "Grapes",
    "Nectarines",
    "Strawberries"
]

counts = [5, 3, 4, 2, 6, 7]
````

# 7. Understanding the Dataset

| Fruit | Count |
|---|---|
| Apples | 5 |
| Pears | 3 |
| Plums | 4 |
| Grapes | 2 |
| Nectarines | 6 |
| Strawberries | 7 |

# 8. Key Concept: Categories

Each fruit is a category.

Categories are:

- labels
- groups
- symbolic entities

They are NOT continuous numerical measurements.

# 9. Why `x_range` Matters

This is the most important technical concept in categorical plotting.

When we write:

```python
figure(x_range=fruits)
```

we are telling Bokeh:

```text
"Treat the x-axis as categorical."
```

Without `x_range`, Bokeh expects numerical coordinates.

# 10. Internal Bokeh Architecture

Internally, Bokeh creates:

```text
FactorRange
```

This object manages:

- category ordering
- category spacing
- categorical rendering
- browser layout

## Internal Flow

```mermaid
flowchart TD

A[Categories]
    --> B[x_range]
    --> C[FactorRange]
    --> D[Categorical Axis]
    --> E[Bar Glyph Rendering]
```

# 11. Create Categorical Figure

````python
p = figure(
    x_range=fruits,
    height=350,
    width=700,
    title="Fruit Counts",
    x_axis_label="Fruit Type",
    y_axis_label="Count"
)
````

# 12. Understanding `vbar()`

`vbar()` means:

```text
Vertical Bar Glyph
```

Internally this is still a glyph renderer.

# 13. Add Vertical Bar Glyph

````python
p.vbar(
    x=fruits,
    top=counts,
    width=0.6,
    color="firebrick"
)

show(p)
````

# 14. Understanding `vbar()` Parameters

| Parameter | Meaning |
|---|---|
| x | category labels |
| top | bar heights |
| width | bar thickness |
| color | bar color |

# 15. Understanding `top`

```python
top=counts
```

means:

- the upper edge of the bar

Bars start at:

```text
y = 0
```

and extend upward to the value in `counts`.

# 22. Why Categorical Data Needs Different Handling

Continuous axes support:

- interpolation
- arithmetic operations
- distance measurements

Categories do not.

Example:

```text
Apple - Pear
```

has no mathematical meaning.

# 23. Ordering Categories

Category order strongly affects interpretation.

## Alphabetical Order

```text
Apple
Grape
Pear
```

## Frequency-Based Order

```text
Most common → least common
```

Frequency ordering is often more insightful.

# 24. Sorting Categories by Count

````python
sorted_pairs = sorted(
    zip(fruits, counts),
    key=lambda x: x[1],
    reverse=True
)

sorted_fruits = [item[0] for item in sorted_pairs]
sorted_counts = [item[1] for item in sorted_pairs]
````

# 25. Plot Sorted Bar Chart

````python
sorted_plot = figure(
    x_range=sorted_fruits,
    height=350,
    width=700,
    title="Sorted Fruit Counts",
    x_axis_label="Fruit",
    y_axis_label="Count"
)

sorted_plot.vbar(
    x=sorted_fruits,
    top=sorted_counts,
    width=0.6,
    color="navy"
)

show(sorted_plot)
````

# 28. Common Beginner Mistakes

## Forgetting `x_range`

Bokeh may interpret categories incorrectly.

## Mismatched Lengths

Wrong:

```python
fruits = ["A", "B"]
counts = [1]
```

## Too Many Categories

100+ categories become unreadable.

Alternatives:

- filtering
- grouping
- treemaps
- aggregation

## Random Color Usage

Use color intentionally.

# 29. Performance Considerations

Bar charts are generally lightweight.

However, extremely large category counts can create:

- browser rendering overhead
- label collisions
- slow interactivity

Solutions:

- pagination
- filtering
- aggregation
- hierarchical grouping
