# Bokeh Glyphs, Widgets, and Layouts

## Introduction

Bokeh is not merely a plotting library.

It is an interactive visualization framework designed around:

- Browser-native rendering
- Interactive exploration
- Dynamic controls
- Dashboard architecture
- Analytical storytelling

Core pipeline:

```text
Figure
  ↓
Glyphs
  ↓
Interactivity
  ↓
Widgets
  ↓
Layouts
  ↓
Interactive Dashboard
```


## Bokeh Visualization Architecture

```mermaid
flowchart TD

A[Data]
    --> B[Figure Canvas]
    --> C[Glyph Rendering]
    --> D[Interactive Tools]
    --> E[Widgets]
    --> F[Layouts]
    --> G[Interactive Dashboard]
```

## Import Required Libraries

````python
import numpy as np
import pandas as pd

from datetime import date

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import DateRangeSlider
from bokeh.layouts import row, column
````

## Initialize Notebook Rendering

````python
output_notebook()
````

## Why `output_notebook()` Matters

Bokeh renders visualizations through JavaScript inside the browser.

Without initialization, interactive plots may fail to render correctly inside notebook environments.

```mermaid
flowchart LR

A[Notebook Environment]
    --> B[output_notebook()]
    --> C[Load BokehJS]
    --> D[Enable Interactive Rendering]
```

# 1. Figure Canvas

The figure canvas defines:

- Coordinate system
- Plot boundaries
- Toolbar tools
- Width and height
- Axis configuration
- Rendering engine


## Generate Data

````python
x = list(range(1, 51))

# Parabola function
y = [i**2 for i in x]
````

## Understanding Function Mapping

The plotted function is:

```text
y = x²
```

Example:

| x | y |
|---|---|
| 1 | 1 |
| 2 | 4 |
| 3 | 9 |
| 4 | 16 |
| 5 | 25 |


## Create Figure Canvas

````python
p = figure(
    title='Parabola Plot',
    x_axis_label='X Values',
    y_axis_label='Y Values',
    width=700,
    height=400,
    tools='pan,wheel_zoom,box_zoom,reset,save'
)
````

## Built-In Interactive Tools

| Tool | Purpose |
|---|---|
| pan | Move chart across canvas |
| wheel_zoom | Mouse wheel zoom |
| box_zoom | Region-based zoom |
| reset | Restore original state |
| save | Export visualization |


# 2. Glyphs

Glyphs are visual representations of data.

Examples:

- line
- circle
- square
- bar
- triangle
- patch
- scatter


## Glyph Mental Model

```text
Data
  ↓
Glyph
  ↓
Visual Representation
```

## Add Line Glyph

````python
p.line(
    x,
    y,
    line_width=3,
    color='blue',
    legend_label='y = x²'
)
````

## Render Visualization

````python
show(p)
````

## Internal Rendering Pipeline

```mermaid
flowchart TD

A[Data]
    --> B[Figure]
    --> C[Glyph Layer]
    --> D[HTML + JavaScript]
    --> E[Browser Rendering]
```

# 3. Multiple Glyph Layers

Modern analytical charts combine multiple visual encodings.

This allows:

- Trend overlays
- Observation markers
- Forecast comparison
- Anomaly highlighting
- Multi-layer storytelling


## Generate Multi-Glyph Data

````python
x_base = [1, 2, 3, 4]
y_base = [2, 5, 4, 6]

# Offset dataset
y_offset = [val - 1 for val in y_base]
````

## Create Layered Figure

````python
p1 = figure(
    title='Layered Glyph Architecture',
    x_axis_label='X Values',
    y_axis_label='Y Values',
    width=700,
    height=400,
    tools='pan,box_zoom,wheel_zoom,reset,save'
)
````

## Add Line Glyph

````python
p1.line(
    x_base,
    y_base,
    legend_label='Trendline',
    line_width=3,
    line_color='navy'
)
````

## Add Circle Glyph

````python
p1.circle(
    x_base,
    y_base,
    legend_label='Circle Markers',
    size=10,
    fill_color='red',
    line_color='black'
)
````

## Add Square Glyph

````python
p1.square(
    x_base,
    y_offset,
    legend_label='Offset Baseline',
    size=12,
    fill_color='orange',
    line_color='darkorange'
)
````

## Multi-Glyph Architecture

```mermaid
flowchart TD

A[Figure Canvas]
    --> B[Line Glyph]
    --> C[Circle Glyph]
    --> D[Square Glyph]

B --> E[Composite Visualization]
C --> E
D --> E
```

# 4. Clickable Legends

Legends in Bokeh are not passive labels.

They can become interactive control systems.

## Activate Interactive Legend Policy

````python
p1.legend.title = 'Interactive Layers'

p1.legend.click_policy = 'hide'
````

## Why Clickable Legends Matter

Users can:

- Hide clutter
- Isolate trends
- Compare categories
- Reduce visual overload
- Explore independently


## Interactive Legend Pipeline

```mermaid
flowchart LR

A[User Clicks Legend]
    --> B[BokehJS Event]
    --> C[Toggle Glyph Visibility]
    --> D[Updated Visualization]
```

## Render Interactive Layered Plot

````python
show(p1)
````

## Create DateRangeSlider Widget

````python
date_slider = DateRangeSlider(
    title='Temporal Filter Range',
    start=date(2022, 1, 1),
    end=date(2023, 12, 31),
    value=(date(2022, 10, 1), date(2022, 12, 31)),
    step=1,
    width=600
)
````

# 6. Layouts

Real-world dashboards rarely contain a single chart.

Layouts help organize:

- Multiple charts
- KPI panels
- Controls
- Dashboards
- Comparative analytics


## Generate Layout Data

````python
x_layout = list(range(1, 11))

y0 = [x for x in x_layout]
y1 = [10 - x for x in x_layout]
y2 = [abs(x - 5) for x in x_layout]
````

## Create Uniform Figure Settings

````python
fig_settings = dict(
    width=260,
    height=260,
    tools='pan,wheel_zoom,reset'
)
````

## Create Plot 1

````python
s1 = figure(
    title='Linear Function (y=x)',
    background_fill_color='#fafafa',
    **fig_settings
)

s1.scatter(
    x_layout,
    y0,
    size=8,
    color='teal',
    marker='circle'
)
````

## Create Plot 2

````python
s2 = figure(
    title='Inverse Function (y=10-x)',
    background_fill_color='#f0f0f0',
    **fig_settings
)

s2.scatter(
    x_layout,
    y1,
    size=8,
    color='crimson',
    marker='square'
)
````

## Create Plot 3

````python
s3 = figure(
    title='Absolute Function (y=|x-5|)',
    background_fill_color='#e6e6e6',
    **fig_settings
)

s3.scatter(
    x_layout,
    y2,
    size=8,
    color='indigo',
    marker='triangle'
)
````

## Horizontal Row Layout

````python
horizontal_dashboard = row(s1, s2, s3)

show(horizontal_dashboard)
````

## Row Layout Mental Model

```text
[Plot 1] [Plot 2] [Plot 3]
```

## Vertical Column Layout

````python
vertical_dashboard = column(s1, s2, s3)

show(vertical_dashboard)
````

## Column Layout Mental Model

```text
[Plot 1]
[Plot 2]
[Plot 3]
```
