# Bokeh Interactive Visualization

## Introduction

Bokeh is a Python library for creating interactive, browser-based visualizations.

Unlike static visualization libraries such as Matplotlib and Seaborn, Bokeh generates:

- HTML
- JavaScript
- Interactive browser-rendered visualizations

This enables:

- Zooming
- Panning
- Hover interactions
- Live updates
- Dashboards
- Web deployment


## Storytelling Philosophy

Bokeh supports two major storytelling paradigms:

### Author-Driven Narrative

The creator defines:

- Initial visualization
- Layout
- Key message
- Baseline interpretation

### Audience-Driven Exploration

The audience can:

- Zoom into data
- Hover over observations
- Pan across charts
- Filter data dynamically
- Investigate patterns independently

This transforms visualization from static storytelling into interactive exploration.

## Import Required Libraries

````python
import numpy as np
import pandas as pd

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import HoverTool, ColumnDataSource
````

## Initializing Notebook Rendering

This is one of the most important steps in Bokeh.

`output_notebook()` injects the required JavaScript rendering environment into Jupyter or Colab notebooks.

````python
output_notebook()
````

## Generate Sample Data

````python
np.random.seed(42)

x_data = np.linspace(0, 10, 50)
y_data = np.sin(x_data) + np.random.normal(0, 0.1, 50)
````

## Creating Your First Bokeh Figure

````python
p = figure(
    title='Interactive Wave Pattern',
    x_axis_label='Time Interval',
    y_axis_label='Signal Value',
    width=700,
    height=400,
    tools='pan,box_zoom,wheel_zoom,reset,save,hover'
)
````

## Understanding Figure Objects

`figure()` creates the plotting canvas.

It defines:

- Width
- Height
- Axis labels
- Interactive tools
- Rendering space


## Adding Glyphs

Glyphs are visual objects rendered onto the canvas.

Examples:

- line
- circle
- rectangle
- patch
- vbar
- scatter


````python
# Add continuous line
p.line(
    x_data,
    y_data,
    legend_label='Trend',
    line_width=2,
    line_color='navy'
)

# Add scatter points
p.circle(
    x_data,
    y_data,
    legend_label='Observations',
    size=8,
    color='orange',
    alpha=0.7
)
````

## Styling the Visualization

````python
p.legend.location = 'top_right'
p.title.text_font_size = '14pt'
````

## Rendering the Plot

````python
show(p)
````

## Adding Hover Interactivity

````python
hover_plot = figure(
    title='Hover Tool Example',
    width=700,
    height=400,
    tools='pan,wheel_zoom,box_zoom,reset'
)

hover_plot.circle(x_data, y_data, size=12)

hover_tool = HoverTool(
    tooltips=[
        ('X Value', '@x'),
        ('Y Value', '@y')
    ]
)

hover_plot.add_tools(hover_tool)

show(hover_plot)
````

## ColumnDataSource

This is the central synchronized data structure in Bokeh.

It acts as a bridge between:

- Python backend
- Browser frontend


````python
source = ColumnDataSource(data={
    'x': [1, 2, 3, 4],
    'y': [10, 15, 8, 20]
})

source.data
````

## ColumnDataSource Architecture

```mermaid
flowchart LR

    --> B[ColumnDataSource]

    --> D[Browser Visualization]
```

## Interactive Scatter Plot Using ColumnDataSource

````python
interactive_plot = figure(
    title='Interactive Scatter Plot',
    width=700,
    height=400,
    tools='pan,wheel_zoom,box_zoom,reset,hover'
)

interactive_plot.circle(
    'x',
    'y',
    source=source,
    size=15,
    color='green'
)

show(interactive_plot)
````

## Streaming Data Example

````python
from bokeh.plotting import curdoc
from bokeh.driving import linear
import random

stream_source = ColumnDataSource(data=dict(x=[], y=[]))

stream_plot = figure(width=700, height=300)

stream_plot.line('x', 'y', source=stream_source)

@linear()
def update(step):
    new_data = {
        'x': [step],
        'y': [random.randint(0, 100)]
    }

    stream_source.stream(new_data, rollover=50)

curdoc().add_periodic_callback(update, 1000)
curdoc().add_root(stream_plot)
````

## Minimal End-to-End Example

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 2, 7]

plot = figure(
    title='My First Bokeh Plot',
    x_axis_label='X Axis',
    y_axis_label='Y Axis',
    width=700,
    height=400,
    tools='pan,wheel_zoom,box_zoom,reset,save'
)

plot.line(x, y, line_width=2)

plot.circle(x, y, size=10)

show(plot)
````
