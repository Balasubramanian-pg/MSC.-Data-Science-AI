# Getting Started with Matplotlib

This notebook transitions from visualization theory into practical implementation using the Python visualization library **Matplotlib**. 

The goal is simple:
- Generate a sine wave
- Understand plotting workflow
- Introduce basic visualization mechanics in Python

### Core Workflow of Visualization in Python
1. Import Libraries
2. Prepare Data
3. Generate Mathematical Values
4. Plot Visualization
5. Render Graph

## 1. Importing Libraries
Python itself contains minimal built-in visualization capability. We need `matplotlib.pyplot` for our plotting interface and `numpy` for arrays and mathematical operations.

```python
import matplotlib.pyplot as plt
import numpy as np

# This command allows plots to be displayed directly inside the notebook
%matplotlib inline
```

## 2. Creating Data & Generating a Sine Wave

We will use `np.linspace(start, stop, num_points)` to generate evenly spaced numerical values over an interval. Smooth plots require dense sampling.

Then, we compute the sine function efficiently using NumPy's vectorized computation.

```python
# Create 100 evenly spaced x values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute sine values simultaneously without explicit loops (Vectorization)
y = np.sin(x)

# Create the line graph
plt.plot(x, y)

# Render the figure visually
plt.show()
```

## 3. Building and Customizing Plots: Multiple Functions

Plotting is not just drawing; it is controlled visual communication. We will now move to an Object-Oriented plotting pattern using `plt.subplots()` which separates the **Figure** (the outer container) from the **Axes** (the actual graphing region).

We will plot both a sine wave and a cosine wave.

```python
# Generate cosine values
y2 = np.cos(x)

# Create figure and axes
fig, ax = plt.subplots()

# Plot the sine wave (Solid blue line)
ax.plot(x, y, color='blue', linestyle='-', label='sine')

# Plot the cosine wave (Dashed red line)
ax.plot(x, y2, color='red', linestyle='--', label='cosine')

# Add chart title
ax.set_title("Sine and Cosine Waves")

# Add axis labels
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Add grid and legend
ax.grid(True)
ax.legend()

# Display plot
plt.show()
```

## 4. Plotting a Linear Equation

Let's apply the plotting pipeline to visualize a straight-line equation: `y = 2x + 5`.
This connects algebra, arrays, and rendering geometry.

```python
# Generate x values from -5 to 5
x_lin = np.linspace(-5, 5, 100)

# Define linear equation
y_lin = 2 * x_lin + 5

# Create figure and axes
fig, ax = plt.subplots()

# Plot the line with customization
ax.plot(x_lin, y_lin, color='green', linestyle='-.', label='y = 2x + 5')

# Add title and labels
ax.set_title("Linear Function")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Add grid and legend
ax.grid(True)
ax.legend()

# Show plot
plt.show()
```

## 5. Building Bar Charts

Bar charts are among the most important business visuals because humans compare lengths very effectively. They are highly effective for categorical analysis and ranking.

Here, we map categorical arrays to value arrays.

```python
# Define categories and values
fruits = ['Apple', 'Blueberry', 'Cherry', 'Orange']
counts = [40, 100, 30, 55]

# Define colors for visual distinction (pre-attentive processing)
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# Create figure and axes
fig, ax = plt.subplots()

# Create customized bar chart
ax.bar(fruits, counts, color=bar_colors)

# Semantic Labeling
ax.set_title("Food Supply by Kind and Color")
ax.set_xlabel("Fruit")
ax.set_ylabel("Count")

# Show plot
plt.show()
```
