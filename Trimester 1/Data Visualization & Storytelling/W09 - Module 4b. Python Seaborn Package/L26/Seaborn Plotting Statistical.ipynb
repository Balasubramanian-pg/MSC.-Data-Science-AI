## Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn theme
sns.set_theme(style='darkgrid')
```

## Understanding Scatter Plots

Scatter plots are used to understand relationships between two variables.

They help identify:

- Positive relationships
- Negative relationships
- No relationship
- Clusters
- Outliers


## Load the Tips Dataset

```python
tips = sns.load_dataset('tips')

tips.head(10)
```

### Dataset Columns

- `total_bill`: Total restaurant bill amount
- `tip`: Tip given by customer
- `sex`: Gender of customer
- `smoker`: Smoker or non-smoker
- `day`: Day of visit
- `time`: Lunch or dinner
- `size`: Group size


## Basic Scatter Plot

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip'
)

plt.show()
```

### Observation

There is a positive relationship between total bill amount and tip amount.

As bill amount increases, tip amount generally increases.

## Adding Semantic Information Using Hue

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker'
)

plt.show()
```

## Why Use Hue?

`hue` adds color encoding to represent categories.

This allows us to compare:

- Smokers
- Non-smokers

inside the same visualization.

## Adding Style Along with Hue

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='smoker'
)

plt.show()
```

## Why Style Matters

Using shapes improves visual cognition.

- Colors alone can be hard to distinguish
- Shapes create stronger category separation
- Improves accessibility and readability


## Adding Another Variable with Style

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    style='time'
)

plt.show()
```

## Interpretation

- Color represents smoker/non-smoker
- Shape represents lunch/dinner

This allows multi-dimensional analysis within a single plot.

## Bubble Plot Using Size

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    size='size'
)

plt.show()
```

## Custom Bubble Sizes

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    size='size',
    sizes=(15, 200)
)

plt.show()
```

## Understanding Line Plots

Line plots emphasize continuity and trends over time.

Common use cases:

- Stock prices
- Time series analysis
- Sensor data
- Economic indicators


## Load Dow Jones Dataset

```python
dowjones = sns.load_dataset('dowjones')

dowjones.head()
```

## Basic Line Plot

```python
sns.relplot(
    data=dowjones,
    x='Date',
    y='Price',
    kind='line'
)

plt.xticks(rotation=45)
plt.show()
```

## Confidence Intervals in Seaborn

Seaborn automatically computes:

- Mean trend
- Confidence interval bands

These help understand uncertainty and variability.

## Working with the fMRI Dataset

```python
fmri = sns.load_dataset('fmri')

fmri.head()
```

## Conditional Line Plot

```python
sns.relplot(
    data=fmri.query("event == 'stim'"),
    x='timepoint',
    y='signal',
    hue='region',
    units='subject',
    estimator=None,
    kind='line'
)

plt.show()
```

## Key Concepts

- `query()` filters data
- `units='subject'` separates observations
- `estimator=None` disables averaging
- Raw trends become visible


## Facet Grids for Multi-Variable Relationships

Facet grids divide plots into multiple subplots.

Useful for:

- Comparing categories
- Multi-dimensional analysis
- Understanding interactions between variables


## Facet Grid Example with Lunch and Dinner

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    col='time'
)

plt.show()
```

## Business Insight

Dinner customers tend to:

- Spend more
- Tip more frequently
- Generate larger revenue opportunities

Restaurant managers can use this for:

- Staffing decisions
- Shift optimization
- Incentive allocation


## Multi-Grid Facet Example

```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    hue='subject',
    col='region',
    row='event',
    height=3,
    estimator=None,
    kind='line'
)

plt.show()
```

## Understanding the Grid Structure

- Rows represent event types
- Columns represent brain regions
- Each subplot shows signal behavior
- Hue distinguishes subjects

This allows detailed comparative analysis.
